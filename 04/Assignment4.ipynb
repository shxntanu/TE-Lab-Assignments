```python
import numpy as np   
import pandas as pd    
from tabulate import tabulate
import matplotlib.pyplot as plt 
from IPython.display import display
%matplotlib inline 
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.set_style("whitegrid")

# Plot 1: Countplot for the number of passengers in each class
plt.figure(figsize=(8, 6))
sns.countplot(x="class", data=titanic, palette="Set1")
plt.title("Passenger Count by Class")
plt.show()
```

### Distribution of Ages using a Histogram

```python
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribution of Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic['deck'] = titanic['deck'].ffill()
titanic['deck'][0] = 'C'
```

```python
titanic = titanic.drop('embark_town', axis=1)
```

```python
titanic['embarked'].value_counts()
```

```python
titanic['embarked'] = titanic['embarked'].fillna('S')
```

```python
bins = [0,100,200,300,400,500]
```

```python
plt.figure(figsize=(5,5))
sns.histplot(data=titanic, x='fare', bins=bins)
plt.show()
```
